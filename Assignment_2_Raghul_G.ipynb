```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/House Price India.csv')

df
```

```python
df1 = df.rename(columns={'number of floors': 'floor',"waterfront present":"wavefront","number of views":"views","condition of the house":"condition"})
```

# TASK -3   VISUALIZATION

1) UNIVARIATE ANALYSIS

```python
sns.distplot(df1.Lattitude)
```

```python
sns.displot(df1.Longitude)
```

```python
plt.pie(df1.views.value_counts(),[0.2,0.1,0,0.1,0],labels = ['0','1','2','3','4'],autopct ='%1.1f%%',shadow = True)
plt.title('NUMBER OF VIEWS')
plt.show()
```

```python
sns.barplot(x=df1.floor.value_counts().index,y=df1.floor.value_counts())
plt.title('NUMBER OF FLOORS',color="red")
```

2) BI - VARIATE ANALYSIS

```python
sns.lineplot(x = df1.views,y=df1.condition)
```

```python
sns.scatterplot(x = df1.Lattitude,y=df1.Longitude)
```

3) MULTIVARIATE ANALYSIS

```python
sns.heatmap(df1.corr())
plt.figure(figsize=(50,50))
```
